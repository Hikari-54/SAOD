### Задание 3.
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.
Выбором | Вставкой | Обменом | Шелла | Турнирная | Быстрая сортировка | Пирамидальная

```python
import time
# math для логарифма в сортировке Шелла
import math
import random

# Создаем матрицу с заданным количеством строк(m) и столбцов(n)
m = input('m = ')
n = input('n = ')
m = int(m)
n = int(n)
if m <= 0:
    m = 50
if n <= 0:
    n = 50

min_limit = input('min_limit = ')
max_limit = input('max_limit = ')
min_limit = int(min_limit)
max_limit = int(max_limit)
if min_limit == 0:
    min_limit = -250
if max_limit == 0:
    max_limit = 1000 + 7 
if min_limit > max_limit:
    min_limit = -250
    max_limit = 1000 + 7 

# Генерация матрицы из чисел заданного диапазона
Matrix = [[random.randint(min_limit, max_limit) for j in range(n)] for i in range(m)]
print('Матрица:')
# Для каждой сортировки копируем изначальную матрицу
MatrixSel=MatrixIn=MatrixPop=MatrixShell=MatrixTour=MatrixQuick=MatrixPyr=Matrix
for i in range(m):
    print(Matrix[i])
    

# Сортировка выбором ----- (В среднем O(n²)) -------------------

def selection_sort(array):
    # В строке массива принимаем первый эл-нт за минимальный
    for i in range(len(array)):
        minimal = i
        # Затем его сравниваем со всеми последующими эл-ми
        # Ищем минимальный во всей строке
        for j in range(i + 1, len(array)):
            if array[j] < array[minimal]:
                minimal = j
        # Затем меняем его местами с тем элементом, который стоит первым
        # Отсортированная часть массива - слева, пополняется возрастающими минимальными эл-ми
        array[i], array[minimal] = array[minimal], array[i]
        
start_time = time.time()        
print('')
print('Сортировка выбором:')
# Сортируем по одной строке матрицы
for i in range(m):
    # Для каждой строки вызываем ф-ию сортировки
    selection_sort(MatrixSel[i])
    # И сразу выводим отсортированную строку
    print(MatrixSel[i])
print('')
stop_time = time.time()
# Вычисляем время работы данной сортировки
operating_time = stop_time - start_time
print(str(operating_time) + " секунд")    
print('')


# Сортировка вставкой ----- (В среднем O(n²)) -------------------
# Отсортированная часть строки - слева. Берем следующий эл-нт и сравниваем со всеми предыдущими
# в сортированной части, так находим для него подходящее место.

def insertion_Sort(array):
    for i in range(1, len(array)):
        # Шаг идет в обратном порядке от индекса 'i'
        for j in range(i, 0, -1):
            # Если предыдущий элемент Больше следующего - меняем их местами
            if array[j] < array[j-1]:
                array[j], array[j-1] = array[j-1], array[j]
            else:
                break
    return array

start_time = time.time()
print('Сортировка вставкой:')
for i in range(m):
    insertion_Sort(MatrixIn[i])
    print(MatrixIn[i])
print('')
stop_time = time.time()
# Вычисляем время работы данной сортировки
operating_time = stop_time - start_time
print(str(operating_time) + " секунд")   
print('')


# Сортировка пузырьком ----- (В среднем O(n²)) -------------------
# Сортированая часть строки - справа. Туда "всплывают" максимальные эл-ты.

def bubble_Sort(array):
    # len(array)-1, чтобы у эл-та была пара справа
    for i in range(0, len(array)-1):
        # (len(array)-1-i) -i потому, что отсортированная часть увеличивается справа.
        for j in range(0, len(array)-1-i):
            # если предыдущий эл-нт Больше следующего - меняем их местами
            if array[j] > array[j+1]:
                array[j], array[j+1] = array[j+1], array[j]

start_time = time.time()
print('Сортировка пузырьком:')
for i in range(m):
    bubble_Sort(MatrixPop[i])
    print(MatrixPop[i])
print('')
stop_time = time.time()
# Вычисляем время работы данной сортировки
operating_time = stop_time - start_time
print(str(operating_time) + " секунд")    
print('')


# Сортировка Шелла ----- (В среднем O(n²)) -------------------
# Усовершенствованный алгоритм сортировки вставками. Суть в том, что сравниваются
# не только соседние эл-ты, но и те, которые находятся на расстоянии друг от друга.
# Это расстояние постепенно сокращается и на последней итерации сравниваются соседние
# эл-ты(классическая сортировка вставками).

def shell_Sort(array):
    n = len(array)
    k = int(math.log2(n))
    # Рассчитываем интервал, между которым будут сравниваться эл-ты
    interval = 2**k -1
    while interval > 0:
        for i in range(interval, n):
            temp = array[i]
            j = i
            while j >= interval and array[j - interval] > temp:
                array[j] = array[j - interval]
                j -= interval
            array[j] = temp
        # Уменьшаем интервал сравнения по степеням двойки
        k -= 1
        interval = 2**k -1
    return array

start_time = time.time()
print('Сортировка Шелла:')
for i in range(m):
    shell_Sort(MatrixShell[i])
    print(MatrixShell[i])
print('')
stop_time = time.time()
# Вычисляем время работы данной сортировки
operating_time = stop_time - start_time
print(str(operating_time) + " секунд")    
print('')


# Турнирная сортировка ----- (В среднем O(n*log^2 n)) -------------------
# Организуется небольшая куча (MIN heap), в которой в корень "всплывают" минимальные эл-ты
# Затем эти мин эл-ты переносятся поочередно в список победителей.
# Проигравшие - это эл-ты, которые попали в корень дерева, но уже не могут быть добавлены
# в конец списка победителей, тк их место уже занято и придется сдвигать все эл-ты, чтобы
# поместить их на свое место. Поэтому они помещаются в отдельный список проигравших
# В конце эти два списка сливаются воедино, образуя упорядоченную строку.

def merge_Sort(nums):
    if len(nums)==1:
        return nums
    mid = (len(nums)-1) // 2
    # Создаем два вспомогательных массива, один для победителей, другой для проигравших
    lst1 = merge_Sort(nums[:mid+1])
    lst2 = merge_Sort(nums[mid+1:])
    # Слияение <Победителей> и <Проигравших> в один список
    result = merge(lst1, lst2)
    return result

def merge(lst1, lst2):
    lst = []
    i = 0
    j = 0
    while(i<=len(lst1)-1 and j<=len(lst2)-1):
        if lst1[i]<lst2[j]:
            lst.append(lst1[i])
            i+=1
        else:
            lst.append(lst2[j])
            j+=1
    if i>len(lst1)-1:
        while(j<=len(lst2)-1):
            lst.append(lst2[j])
            j+=1
    else:
        while(i<=len(lst1)-1):
            lst.append(lst1[i])
            i+=1
    return lst

start_time = time.time()
print('Турнирная сортировка:')
for i in range(m):
    merge_Sort(MatrixTour[i])
    print(MatrixTour[i])
print(' ')
stop_time = time.time()
# Вычисляем время работы данной сортировки
operating_time = stop_time - start_time
print(str(operating_time) + " секунд")    
print('')


# Быстрая сортировка ------ (В среднем O(n*log n)) -------------------
# Выбирается опорный эл-нт в списке (k). Все остальные делятся на 3 группы: меньше, равны, больше.
# Они помещаются в отдельные списки. Для которых затем рекурсивно выполняются те же условия
# до тех пор, пока мы не получим список из 1-го эл-та (он априори отсортирован)
# или пока мы не получим пустой список. 

def quick_Sort(Matrix):
    array = Matrix.copy()
    if len(array) < 2:
        return array
    low = []
    same = []
    high = []
    # Опорный элемент выбирается случайно
    k = array[random.randint(0, len(array) - 1)]
    for item in array:
        if item < k:
            low.append(item)
        elif item == k:
            same.append(item)
        elif item > k:
            high.append(item)
    # Появляется рекурсия
    return quick_Sort(low) + same + quick_Sort(high)
    
start_time = time.time()
print(' ')
print('Быстрая сортировка: ')
for i in range(m):
    print(quick_Sort(MatrixQuick[i]))
print(' ')
stop_time = time.time()
# Вычисляем время работы данной сортировки
operating_time = stop_time - start_time
print(str(operating_time) + " секунд")
print('')


# Пирамидальная сортировка ------ (В среднем O(n*log n)) -------------------
# Строка представляется в виде бинарного дерева(по степеням 2-ки, 1/2/4/8 и тд), которое затем нужно
# привести к упорядоченному виду (Родитель > Потомков). После приведения
# к такому виду это уже куча(max heap). Цикл прошел один раз и в корне дерева (1-ый эл-нт)
# находится максимальное число. Его переносим на последнее место и блокируем.
# Затем повторяем эти действия. Отсортированная часть - справа.

# Формирует из списка - Max Heap(Возрастающую кучу)
def heapify_Sort(array, n, i):
    largest = i
    # l - левый потомок эл-та
    l = 2 * i + 1
    # r - правый потомок эл-та
    r = 2 * i + 2

    # Если левый потомок корня — допустимый индекс, а элемент больше,
    # чем текущий наибольший, обновляем наибольший элемент
    if l < n and array[i] < array[l]:
        largest = l
    # То же самое для правого потомка корня
    if r < n and array[largest] < array[r]:
        largest = r
    # Если наибольший элемент больше не корневой, они меняются местами
    if largest != i:
        array[i], array[largest] = array[largest], array[i]
        heapify_Sort(array, n, largest)
        
def heapSort(array):
    n = len(array)
    # Создаем Max Heap из списка
    for i in range(n // 2, -1, -1):
        heapify_Sort(array, n, i)
    # Перемещаем корень Max Heap в конец списка
    for i in range(n - 1, 0, -1):
        array[i], array[0] = array[0], array[i]
        heapify_Sort(array, i, 0)
    return array

start_time = time.time()
print(' ')
print('Пирамидальная сортировка:')
for i in range(m):
    heapSort(MatrixPyr[i])
    print(MatrixPyr[i])
stop_time = time.time()
# Вычисляем время работы данной сортировки
operating_time = stop_time - start_time
print(str(operating_time) + " секунд")
print(' ')


#Стандартная, встроенная сортировка -------------------

start_time = time.time()
print(' ')
print('Стандартная сортировка:')
for i in range(m):
    sorted(Matrix[i])
    print(Matrix[i])
print(' ')
stop_time = time.time()
# Вычисляем время работы данной сортировки
operating_time = stop_time - start_time
print(str(operating_time) + " секунд")
print(' ')
```
